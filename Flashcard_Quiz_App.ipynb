```python
# Install ipywidgets if needed
!pip install ipywidgets

from IPython.display import display, clear_output
import ipywidgets as widgets

# ----------------------
# Sample flashcards
# ----------------------
flashcards = [
    {"question": "What is Python?", "answer": "A programming language"},
    {"question": "What is 2+2?", "answer": "4"},
    {"question": "What is AI?", "answer": "Artificial Intelligence"}
]

current_index = 0

# ----------------------
# UI Elements
# ----------------------
question_label = widgets.Label()
answer_label = widgets.Label()
show_answer_btn = widgets.Button(description="Show Answer")
prev_btn = widgets.Button(description="Previous")
next_btn = widgets.Button(description="Next")
add_btn = widgets.Button(description="Add Flashcard")
edit_btn = widgets.Button(description="Edit Flashcard")
delete_btn = widgets.Button(description="Delete Flashcard")

# Input widgets for Add/Edit
q_input = widgets.Text(description="Question:")
a_input = widgets.Text(description="Answer:")

# ----------------------
# Functions
# ----------------------
def show_question():
    global current_index
    if flashcards:
        question_label.value = f"Q: {flashcards[current_index]['question']}"
        answer_label.value = ""
    else:
        question_label.value = "No flashcards!"
        answer_label.value = ""

def show_answer(b):
    if flashcards:
        answer_label.value = f"A: {flashcards[current_index]['answer']}"

def next_card(b):
    global current_index
    if current_index < len(flashcards) - 1:
        current_index += 1
        show_question()

def prev_card(b):
    global current_index
    if current_index > 0:
        current_index -= 1
        show_question()

def add_card(b):
    q, a = q_input.value, a_input.value
    if q and a:
        flashcards.append({"question": q, "answer": a})
        q_input.value = ""
        a_input.value = ""
        show_question()

def edit_card(b):
    if flashcards:
        q, a = q_input.value, a_input.value
        if q and a:
            flashcards[current_index] = {"question": q, "answer": a}
            show_question()

def delete_card(b):
    global current_index
    if flashcards:
        flashcards.pop(current_index)
        if current_index >= len(flashcards):
            current_index = max(0, len(flashcards)-1)
        show_question()

# ----------------------
# Link buttons
# ----------------------
show_answer_btn.on_click(show_answer)
next_btn.on_click(next_card)
prev_btn.on_click(prev_card)
add_btn.on_click(add_card)
edit_btn.on_click(edit_card)
delete_btn.on_click(delete_card)

# ----------------------
# Display UI
# ----------------------
display(question_label, answer_label)
display(widgets.HBox([prev_btn, show_answer_btn, next_btn]))
display(widgets.Label("Add / Edit Flashcard:"))
display(q_input, a_input)
display(widgets.HBox([add_btn, edit_btn, delete_btn]))

show_question()
```
